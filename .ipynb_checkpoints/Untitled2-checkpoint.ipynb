```python
import csv
import time
import requests
import re
from bs4 import BeautifulSoup
from urllib.parse import quote

import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys

import time
import random
from urllib.parse import urlparse
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager

def find_official_websites(query):
    """Search Bing using Selenium + XPath and return first 3 official-looking websites."""
    unwanted_domains = [
        "wikipedia.org", "facebook.com", "instagram.com", "linkedin.com",
        "twitter.com", "youtube.com", "yelp.com", "tripadvisor.com"
    ]

    options = Options()
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    try:
        driver.get("https://www.bing.com/")
        time.sleep(2)

        # Search bar
        search_box = driver.find_element(By.XPATH, '//*[@name="q"]')

        # Type character-by-character like a human
        for char in query:
            search_box.send_keys(char)
            time.sleep(random.uniform(0.08, 0.18))

        search_box.send_keys(Keys.RETURN)
        time.sleep(2)

        # Get search result links
        results = driver.find_elements(By.XPATH, '//li[@class="b_algo"]//h2/a')

        official_links = []
        for res in results:
            # Open link in a new tab
            driver.execute_script("window.open(arguments[0], '_blank');", res.get_attribute("href"))
            driver.switch_to.window(driver.window_handles[-1])

            time.sleep(2)  # wait for the page to load

            final_url = driver.current_url  # Get the actual destination URL

            # Close the tab and return to search results
            driver.close()
            driver.switch_to.window(driver.window_handles[0])

            # Filter unwanted domains
            if not any(domain in final_url for domain in unwanted_domains):
                parsed = urlparse(final_url)
                domain = parsed.netloc.lower().replace("www.", "")
                if domain not in official_links:
                    official_links.append(domain)

            if len(official_links) == 3:
                break

        return official_links

    except Exception as e:
        print(f"Error searching {query}: {e}")
        return []
    finally:
        driver.quit()

# Example usage
# print(find_official_websites("Milan Institute"))
# CRM detection patterns (lowercase keys, regex or keyword matches)
CRM_FINGERPRINTS = {
    "Salesforce (Salesforce Education Cloud)": {
        "hosts": ["salesforce.com", "force.com", "lightning.force.com"],
        "paths": ["/services/data", "/sfsites/", "/one/one.app"],
        "keywords": ["education cloud", "salesforce lightning", "salesforce apex"]
    },
    "HubSpot (Hubspot CRM)": {
        "hosts": ["js.hs-scripts.com", "hsadspixel.net", "hscollectedforms.net", "hubspot.com", "hs-analytics.net", "hsappstatic.net"],
        "paths": ["/hs-scripts/", "/hs-analytics/", "/collect"],
        "keywords": ["hubspot tracking code", "hubspotutk"]
    },
    "Leadsquared": {
        "hosts": ["leadsquared.com", "lsmgrt.com", "lmsvc.com"],  # lsmgrt is common in LS assets
        "paths": ["/lsjs/"],
        "keywords": ["leadsquared"]
    },
    "Element 451": {
        "hosts": ["element451.com", "element451.io"],
        "paths": ["/api/element451", "/e451"],
        "keywords": ["element451"]
    },
    "Slate (Technolutions)": {
        "hosts": ["technolutions.net", "technolutions.com"],
        "paths": ["/apply/", "/portal/", "/slate/"],
        "keywords": ["technolutions", "slate portal"]
    },
    "Anthology Reach": {
        "hosts": ["anthology.com", "campusnexus.com", "campusmgmt.com"],
        "paths": ["/reach", "/campusnexus"],
        "keywords": ["anthology reach", "campusnexus recruit"]
    },
    "Ellucian (Ellucian CRM Recruit / Ethos / CRM Advise)": {
        "hosts": ["elluciancloud.com", "ellucian.com"],
        "paths": ["/ethos", "/elluciancrmrecruit", "/crm-advise"],
        "keywords": ["ellucian crm", "ellucian ethos"]
    },
    "Velocify": {
        "hosts": ["velocify.com", "leads360.com", "icemortgagetechnology.com"],
        "paths": ["/api/velocify"],
        "keywords": ["velocify"]
    },
    "Sparkroom (by Digital Media Solutions)": {
        "hosts": ["sparkroom.com", "digitalmediasolutions.com"],
        "paths": ["/sparkroom"],
        "keywords": ["sparkroom"]
    },
    "Edlumina": {
        "hosts": ["edlumina.com"],
        "paths": [],
        "keywords": ["edlumina"]
    },
    "EducationDynamics CRM (EDDY)": {
        "hosts": ["educationdynamics.com", "dmsgroup.com", "dmseducation.com"],
        "paths": [],
        "keywords": ["educationdynamics", "eddy crm"]
    },
    "Meritize CRM Suite": {
        "hosts": ["meritize.com"],
        "paths": [],
        "keywords": ["meritize"]
    },
    "FAME SIS/CRM": {
        "hosts": ["fame.edu"],
        "paths": [],
        "keywords": ["fame sis"]
    },
    "Paradigm CRM": {
        "hosts": ["paradigmcrm", "paradigmedu"],  # vendor patterns vary
        "paths": [],
        "keywords": ["paradigm crm"]
    },
    "ClassReach": {"hosts": ["classreach.com"], "paths": [], "keywords": ["classreach"]},
    "Caylor Solutions CRM Stack": {"hosts": ["caylor-solutions.com", "caylor-solutions"], "paths": [], "keywords": ["caylor solutions"]},
    "Azorus": {"hosts": ["azorus.com"], "paths": [], "keywords": ["azorus"]},
    "Zoho": {"hosts": ["zoho.com", "crm.zoho.com", "zohopublic.com"], "paths": ["/crm/"], "keywords": ["zoho crm"]},
    "TargetX": {"hosts": ["targetx.com"], "paths": [], "keywords": ["targetx"]},
    "Creatrix Campus": {"hosts": ["creatrixcampus.com"], "paths": [], "keywords": ["creatrix"]},
    "Embark Campus": {"hosts": ["embark.com", "embarkcampus.com"], "paths": [], "keywords": ["embark campus"]},
    "AdmitTek": {"hosts": ["admittek"], "paths": [], "keywords": ["admittek"]},
    "Admittor": {"hosts": ["admittor"], "paths": [], "keywords": ["admittor"]},
    "Ascend": {"hosts": ["ascendcrm"], "paths": [], "keywords": ["ascend crm"]},
    "BocaVox": {"hosts": ["bocavox.com"], "paths": [], "keywords": ["bocavox"]},
    "SchoolMint": {"hosts": ["schoolmint.com"], "paths": [], "keywords": ["schoolmint"]},
    "STARS Campus Solutions": {"hosts": ["starscampussolutions.com"], "paths": [], "keywords": ["stars campus solutions"]},
    "Finalsite": {"hosts": ["finalsite.com"], "paths": [], "keywords": ["finalsite"]},
    "FileInvite": {"hosts": ["fileinvite.com"], "paths": [], "keywords": ["fileinvite"]},
    "Wisenet": {"hosts": ["wisenet.co"], "paths": [], "keywords": ["wisenet"]},
    "Shape CRM": {"hosts": ["setshape.com"], "paths": [], "keywords": ["shape crm"]},
    "SuiteCRM": {"hosts": ["suitecrm.com"], "paths": [], "keywords": ["suitecrm"]},
    "Vtiger": {"hosts": ["vtiger.com"], "paths": [], "keywords": ["vtiger"]},
    "OroCRM": {"hosts": ["orocrm.com"], "paths": [], "keywords": ["orocrm"]},
    "YetiForce": {"hosts": ["yetiforce.com"], "paths": [], "keywords": ["yetiforce"]},
    "EspoCRM": {"hosts": ["espocrm.com"], "paths": [], "keywords": ["espocrm"]},
}

# pages where CRMs commonly load widgets/scripts (kept small to avoid slowness)
CANDIDATE_PATHS = ["", "/contact", "/apply", "/admissions", "/request-info"]
def _normalize_url(url: str) -> str:
    if not url.startswith(("http://", "https://")):
        return "http://" + url
    return url

def _collect_indicators(base_html: str, soup: BeautifulSoup):
    """Return sets of hostnames and full URLs from strong-signal elements only."""
    hosts = set()
    urls = set()

    def add(url):
        if not url:
            return
        try:
            parsed = urlparse(url)
            if parsed.scheme in ("http", "https"):
                urls.add(url)
                if parsed.netloc:
                    hosts.add(parsed.netloc.lower())
        except Exception:
            pass

    # Scripts & links & iframes & forms & anchors (href/src/action are strong indicators)
    for tag, attr in (("script", "src"), ("link", "href"), ("img", "src"),
                      ("iframe", "src"), ("a", "href"), ("form", "action")):
        for el in soup.find_all(tag):
            add(el.get(attr))

    # Meta generator/powered-by
    for m in soup.find_all("meta"):
        for key in ("content", "name", "property"):
            val = m.get(key)
            if val:
                # treat meta values as 'keywords' sources
                urls.add(val.lower())

    # Also look for raw URLs inside inline scripts once (basic regex)
    inline_urls = set(re.findall(r"https?://[^\s\"'<>]+", base_html, re.I))
    for u in inline_urls:
        add(u)

    return hosts, urls

def _score_crm(crm_name: str, hosts: set, urls: set, page_text: str) -> tuple[int, str | None]:
    """Return (score, evidence) for the CRM based on indicators."""
    fp = CRM_FINGERPRINTS[crm_name]
    score = 0
    evidence = None

    # Host hits are strong (+3 each, keep one evidence)
    for h in list(hosts):
        for want in fp["hosts"]:
            if h == want or h.endswith("." + want) or want in h:
                score += 3
                evidence = evidence or f"host:{h}"
                break

    # Path/API substrings in collected URLs are medium (+2)
    for u in list(urls):
        for p in fp["paths"]:
            if p and p in u:
                score += 2
                evidence = evidence or f"path:{p} in {u}"
                break

    # Brand keywords in meta/inline (lightweight, +1)
    low = page_text
    for kw in fp["keywords"]:
        if kw and kw.lower() in low:
            score += 1
            evidence = evidence or f"kw:{kw}"
            break

    return score, evidence

def detect_crm_smart(html: str, soup: BeautifulSoup) -> tuple[str, str]:
    """Pick the highest scoring CRM; require minimum confidence to avoid false positives."""
    hosts, urls = _collect_indicators(html, soup)
    text_low = html.lower()

    best = ("Other/Unknown", None, 0)  # (name, evidence, score)

    for crm in CRM_FINGERPRINTS:
        score, ev = _score_crm(crm, hosts, urls, text_low)
        if score > best[2]:
            best = (crm, ev, score)

    # Require a reasonable confidence threshold
    # >=3 usually means at least one vendor host matched
    if best[2] >= 3:
        return best[0], (best[1] or "")
    return "Other/Unknown", ""
    

def scrape_site(url):
    """
    Fetch site (and a few common subpages) and detect CRM using strong-signal fingerprints.
    Returns dict with url, title, crm, evidence.
    """
    headers = {"User-Agent": "Mozilla/5.0"}
    base = _normalize_url(url)
    seen = set()

    aggregate_html = ""
    title = None
    base_host = urlparse(base).netloc

    for path in CANDIDATE_PATHS:
        try:
            target = urljoin(base, path)
            if target in seen:
                continue
            seen.add(target)

            resp = requests.get(target, headers=headers, timeout=12)
            # follow redirects; stop if we are bounced off-site too far
            if resp.status_code >= 400:
                continue

            html = resp.text or ""
            aggregate_html += "\n" + html
            if not title:
                soup0 = BeautifulSoup(html, "html.parser")
                if soup0.title and soup0.title.string:
                    title = soup0.title.string.strip()

        except requests.RequestException:
            continue

    if not aggregate_html:
        print(f"❌ Unable to fetch any content for {url}")
        return {"url": url, "title": title or "No title found", "crm": "Not Found", "evidence": ""}

    soup_all = BeautifulSoup(aggregate_html, "html.parser")
    crm_name, evidence = detect_crm_smart(aggregate_html, soup_all)
    title = title or (soup_all.title.string.strip() if soup_all.title and soup_all.title.string else "No title found")

    print(f"✅ {url} — Title: {title} | CRM: {crm_name} | Evidence: {evidence}")
    return {"url": url, "title": title, "crm": crm_name, "evidence": evidence}
    
def process_csv_and_scrape():
    """Read institute names from temp.csv, find websites, and scrape them."""
    with open("temp.csv", "r", newline="", encoding="utf-8") as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            if row:
                name = row[0].strip()
                print(f"\nSearching website for: {name}")
                site_url = find_official_websites(name)
                if site_url:
                    print(f"Found: {site_url}")
                    scrape_site(site_url)
                else:
                    print("No website found.")

# Run
# process_csv_and_scrape()
```

```python
def process_csv_and_scrape():
    """Read institute names from temp.csv, search for Website, SIS, CRM, LMS, and save results incrementally."""
    output_file = "results.csv"

    # Write header first
    with open(output_file, "w", newline="", encoding="utf-8") as outfile:
        writer = csv.writer(outfile)
        writer.writerow(["College / University Name", "Website", "SIS", "CRM", "LMS", "Source Reference"])

    with open("temp.csv", "r", newline="", encoding="utf-8") as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            if not row:
                continue

            name = row[0].strip()
            print(f"\nProcessing: {name}")

            # 1. Website
            website_links = find_official_websites(name)
            website = website_links[0] if website_links else ""

            # # 2. SIS
            # sis_links = find_official_websites(f"{name} SIS")
            # sis = sis_links[0] if sis_links else ""

            # 3. CRM
            crm_links = find_official_websites(f"{name} CRM")
            crm = crm_links[0] if crm_links else ""

            # # 4. LMS
            # lms_links = find_official_websites(f"{name} LMS")
            # lms = lms_links[0] if lms_links else ""

            # Append row directly to file
            with open(output_file, "a", newline="", encoding="utf-8") as outfile:
                writer = csv.writer(outfile)
                # print(name, website, sis, crm, lms, "\n")
                # writer.writerow([name, website, sis, crm, lms, "Bing Search"])

                print(name, website, crm, "\n")

                scrape_site(crm)
                writer.writerow([name, website, crm, "Bing Search"])

                

            print(f"✅ Saved: {name}")


```
